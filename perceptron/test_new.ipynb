```python
from perceptron_new import generate_boolean_inputs
from perceptron_new import BooleanMLP
import numpy as np
```

```python
X = generate_boolean_inputs(2)
y_xor = np.array([-1, 1, 1, -1])
# y_xor=np.array([-1,1,1,-1,1,-1,-1,1])
y_or=np.array([-1,1,1,1])
y_and=np.array([-1,-1,-1,1])
y_nand=np.array([1,1,1,-1])
y_nor=np.array([1,-1,-1,-1])
print(f"the X for boolean inputs is: {X}")
print(X.shape)
```

```python
mlp_thresh = BooleanMLP(2, mode="threshold")
mlp_thresh.train(X, y_xor)
print("Threshold XOR:", mlp_thresh.predict([-1,1]))
```

```python
mlp_sigmoid = BooleanMLP(2, mode="sigmoid")
mlp_sigmoid.train(X, y_xor)
print("Sigmoid XOR:", mlp_sigmoid.predict(X))
```

```python
mlp_thresh = BooleanMLP(2, mode="threshold")
mlp_thresh.train(X, y_or)
print("Threshold OR:", mlp_thresh.predict(X))
```

```python
mlp_sigmoid = BooleanMLP(2, mode="sigmoid")
mlp_sigmoid.train(X, y_or)
print("Sigmoid OR:", mlp_sigmoid.predict(X))
```

```python
mlp_thresh = BooleanMLP(2, mode="threshold")
mlp_thresh.train(X, y_and)
print("Threshold AND:", mlp_thresh.predict(X))
```

```python
mlp_sigmoid = BooleanMLP(2, mode="sigmoid")
mlp_sigmoid.train(X, y_and)
print("Sigmoid AND:", mlp_sigmoid.predict(X))
```

```python
mlp_thresh = BooleanMLP(2, mode="threshold")
mlp_thresh.train(X, y_nand)
print("Threshold NAND:", mlp_thresh.predict(X))
```

```python
mlp_sigmoid = BooleanMLP(2, mode="sigmoid")
mlp_sigmoid.train(X, y_nand)
print("Sigmoid NAND:", mlp_sigmoid.predict(X))
```
