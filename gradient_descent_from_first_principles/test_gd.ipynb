```python
import numpy as np
from gradient_descent import GradientDescent
import matplotlib.pyplot as plt
```

```python
learning_rates=[0.01,0.1,0.5]
parameters=[[2,2],[2,2],[2,2]]
```

```python
def f_surface(params):
    x,y=params
    return x**2 +y**2 + 10*np.sin(x) + 10*np.cos(y)
```

```python
for lr,params in zip(learning_rates,parameters):
     x = np.linspace(-5, 5, 100)
     y = np.linspace(-5, 5, 100)
     X, Y = np.meshgrid(x, y)
     Z = f_surface([X, Y])

     gd = GradientDescent(lr=lr, max_iters=100)
     positions, costs = gd.step(params)

    
     plt.figure(figsize=(12, 5))

     plt.subplot(1, 2, 1)
     plt.contourf(X, Y, Z, levels=20, cmap='viridis', alpha=0.8)
     plt.contour(X, Y, Z, levels=20, colors='black', linewidths=0.3)
     path_x, path_y = positions[:, 0], positions[:, 1]

     plt.plot(path_x, path_y, 'ro-', linewidth=3, markersize=8, 
          markerfacecolor='red', markeredgecolor='darkred', markeredgewidth=1.5,
          label=f'GD Path (lr={lr}, start={params})')

     plt.plot(path_x[0], path_y[0], 'go', markersize=15, markerfacecolor='green', 
          markeredgecolor='darkgreen', markeredgewidth=2, label='Start')
     plt.plot(path_x[-1], path_y[-1], 'bs', markersize=15, markerfacecolor='blue', 
          markeredgecolor='darkblue', markeredgewidth=2, label='Minimum')

     plt.xlabel('x', fontsize=12)
     plt.ylabel('y', fontsize=12)
     plt.title('GD Trajectory on Contour', fontsize=14)
     plt.legend()
     plt.colorbar(label='f(x,y)')

     plt.subplot(1, 2, 2)
     plt.semilogy(costs, 'go-', linewidth=2, markersize=6)
     plt.xlabel('Iteration', fontsize=12)
     plt.ylabel('Cost(log scale)', fontsize=12)
     plt.title('Cost Convergence', fontsize=14)
     plt.grid(True, alpha=0.3)

     plt.tight_layout()
     plt.show()

     print(f"Final position: {positions[-1]}")
     print(f"Final cost: {costs[-1]:.6f}")
    
```
