```python
from neural_net import NeuralNet
from loss import CrossEntropy
from optim import SGD
epochs=100
model=NeuralNet(input_size=784,output_size=10)

optimizer=SGD([model.fc1,model.fc2],lr=0.01)
```

```python
for epoch in range(epochs):
    out=model.forward(X)
    Criterion=CrossEntropy(out,y)
    dout=Criterion.backward()
    print(dout.shape)
    model.backward(dout)

    optimizer.step()

    optimizer.zero_grad()
    if epoch%10==0:
        print(f"for epoch {epoch+1} ,the loss is {Criterion.item():.4f}")









    
    

```
